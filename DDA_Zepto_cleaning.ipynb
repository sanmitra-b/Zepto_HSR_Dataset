```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
df = pd.read_excel('/content/zepto_bangalore_all_stores_20260219_1358.xlsx')
df
```

```python
# Get the current number of rows in the DataFrame
num_rows = len(df)

# Define the first segment (rows 1 to 6455, 0-indexed: 0 to 6454)
segment1_end_idx = 6455 # Exclusive index, so goes up to index 6454
df_segment1 = df.iloc[0:segment1_end_idx]

# Define the second segment (rows 6457 to end, 0-indexed: 6456 to num_rows-1)
segment2_start_idx = 6456 # Inclusive index
df_segment2 = df.iloc[segment2_start_idx:]

print(f"Number of rows in segment 1 (1 to 6455): {len(df_segment1)}")
print(f"Number of rows in segment 2 (6457 to end): {len(df_segment2)}")

# Perform an inner merge to find rows that are common (exact duplicates) in both segments
common_rows_df = pd.merge(df_segment1, df_segment2, how='inner')

if not common_rows_df.empty:
    print(f"\nThere are {len(common_rows_df)} rows in the second segment (rows 6457 to end) that are exact duplicates of rows in the first segment (rows 1 to 6455).")
    print("Here are the first 5 examples of such common rows:")
    display(common_rows_df.head())
else:
    print("\nThere are no rows in the second segment (rows 6457 to end) that are exact duplicates of rows in the first segment (rows 1 to 6455).")

# Check if the second segment is a complete subset of the first segment
# This means every row in df_segment2 must also exist in df_segment1
is_subset = len(common_rows_df) == len(df_segment2) and len(df_segment2) > 0

if is_subset:
    print("\nAll rows in the second segment (rows 6457 to end) are duplicates of rows in the first segment (rows 1 to 6455).")
else:
    print("\nNot all rows in the second segment (rows 6457 to end) are duplicates of rows in the first segment.")
```

```python
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
#For No ratings Fill Rating = 0
df["Average Rating"] = df["Average Rating"].fillna(0)
df["Total Ratings"] = df["Total Ratings"].fillna(0)
```

```python
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
df = df.drop(columns=['L3 Category'])
print("Dropped 'L3 Category' column due to high number of nulls.")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
#Create Label Column (High Discount Classification)

df["High_Discount_Flag"] = np.where(
    df["Discount %"] >= 20,
    True,
    False
)

print("Products with High Discount:")
display(df[df["High_Discount_Flag"] == True][["Product Name", "Discount %", "Selling Price (₹)", "High_Discount_Flag"]].head())
```

```python
top_10_discount_categories = df.groupby('Primary Category')['Discount %'].mean().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Discount %', y='Primary Category', data=top_10_discount_categories, palette='viridis', hue='Primary Category', legend=False)
plt.title('Top 10 Primary Categories with Highest Average Discounts')
plt.xlabel('Average Discount %')
plt.ylabel('Primary Category')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount %', y='Average Rating', data=df[df['Average Rating'] > 0], alpha=0.6)
plt.title('Discount % vs. Average Rating (Excluding products with no ratings)')
plt.xlabel('Discount %')
plt.ylabel('Average Rating')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Identify duplicate product variants for inspection
duplicate_variants = df[df.duplicated(subset=["Product Variant ID"], keep=False)]

if not duplicate_variants.empty:
    print("Found duplicate product variants. Here are the first 5 examples:")
    display(duplicate_variants.sort_values(by="Product Variant ID").head())
else:
    print("No duplicate product variants found.")

```

```python
# Check if products with Available Quantity = 0 are incorrectly marked as In Stock
inconsistent_stock = df[(df["Available Quantity"] == 0) & (df["Out of Stock"] == False)]

if not inconsistent_stock.empty:
    print("Found inconsistencies where 'Available Quantity' is 0 but 'Out of Stock' is False:")
    display(inconsistent_stock.head())
else:
    print("No inconsistencies found: All products with 0 'Available Quantity' are correctly marked as 'Out of Stock'.")
```

```python
print("Descriptive statistics for 'Total Ratings':")
display(df['Total Ratings'].describe())

print("\nDescriptive statistics for 'Average Rating':")
display(df['Average Rating'].describe())
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['Total Ratings'], bins=50, kde=True)
plt.title('Distribution of Total Ratings')
plt.xlabel('Total Ratings')
plt.ylabel('Number of Products')

plt.subplot(1, 2, 2)
sns.histplot(df['Average Rating'], bins=20, kde=True)
plt.title('Distribution of Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Number of Products')

plt.tight_layout()
plt.show()
```

```python
high_demand_products = df[(df['Total Ratings'] > 5000) & (df['Average Rating'] >= 4.5)]

print(f"Found {len(high_demand_products)} products classified as high demand.\n")

print("High-demand products by Primary Category:")
display(high_demand_products['Primary Category'].value_counts().reset_index(name='Number of High Demand Products'))
```

```python
# Initialize High_Demand_Flag to False for all products
df['High_Demand_Flag'] = False

# Set High_Demand_Flag to True for products identified as high demand
df.loc[df['Product Variant ID'].isin(high_demand_products['Product Variant ID']), 'High_Demand_Flag'] = True

print("First 5 products marked as high demand with the new flag:")
display(df[df['High_Demand_Flag'] == True][['Product Name', 'Total Ratings', 'Average Rating', 'High_Demand_Flag']].head())
```
